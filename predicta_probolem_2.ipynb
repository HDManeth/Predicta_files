```python
import pandas as pd
import numpy as np
```

```python
all_data = pd.read_csv('/content/daily_data.csv')
all_data
```

```python
from sklearn.preprocessing import LabelEncoder
label_encode = LabelEncoder()
all_data['city_id'] = label_encode.fit_transform(all_data['city_id'])
all_data
```

```python
all_data['sunrise'] = pd.to_datetime(all_data['sunrise'], format='%I:%M %p')
```

```python
all_data['sunset'] = pd.to_datetime(all_data['sunset'], format='%I:%M %p')
```

```python
reference_datetime = pd.to_datetime("1900-01-01 00:00:00")
all_data['sunrise'] = (all_data['sunrise'] - reference_datetime).dt.total_seconds()
```

```python
reference_datetime1 = pd.to_datetime("1900-01-01 12:00:00")
all_data['sunset'] = (all_data['sunset'] - reference_datetime1).dt.total_seconds()
```

```python
all_data
```

```python
# Assuming your original DataFrame is named 'df'
df_nonull = all_data.dropna(how='any')  # Drop rows with missing values
dropped_df = all_data[~all_data.index.isin(df_nonull.index)]  # Create a new DataFrame with dropped rows
```

##This model had the highest accuracy

```python
x = df_nonull.drop(['day_id','condition_text', 'wind_degree', 'pressure_mb', 'air_quality_us-epa-index', 'city_id'], axis=1)
y = df_nonull['condition_text']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=250,max_depth = 200, random_state=38)
model.fit(x_train, y_train)
y_hat = model.predict(x_test)
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_hat)
print(f"Accuracy: {accuracy:.4f}")


```

```python
features = ['city_id', 'temperature_celsius',
       'wind_kph', 'wind_degree', 'pressure_mb', 'precip_mm', 'humidity',
       'cloud', 'feels_like_celsius', 'visibility_km', 'uv_index', 'gust_kph',
       'air_quality_us-epa-index', 'sunrise', 'sunset']
df_nonull2 = df_nonull.drop(['day_id','condition_text'], axis=1)
all_combinations = []
for r in range(2, len(features) + 1):
    combinations_r = combinations(features, r)
    all_combinations.extend(combinations_r)

listofall = []
for combo in all_combinations:
    listofall.append(list(combo))

listofall = listofall[:-1]

listofall_2 = [sublist for sublist in  listofall if len(sublist) == 2]

accuracy = []
for item in listofall_2:
    x = df_nonull2.drop(item, axis=1)
    y = df_nonull['condition_text']
    from sklearn.model_selection import train_test_split
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)

    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier(n_estimators=20,max_depth = 200, random_state=1)
    model.fit(x_train, y_train)
    y_hat = model.predict(x_test)
    from sklearn.metrics import accuracy_score

    accuracy.append(accuracy_score(y_test, y_hat))


MAX = max(accuracy)
print(accuracy.index(MAX))
```
